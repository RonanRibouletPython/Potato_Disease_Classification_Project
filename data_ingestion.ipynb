```python
# Librairies imports
import tensorflow as tf
import matplotlib.pyplot as plt
```

```python
# Hyperparameters
IMAGE_SIZE = 256
BATCH_SIZE = 32
VAL_SPLIT = 0.1
TRAIN_SPLIT = 0.2
```

Import the datasets and change it into tensorflow dataset object:
- Source of the dataset: https://www.kaggle.com/datasets/faysalmiah1721758/potato-dataset

```python
data_dir = "Datasets"
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir,
    seed = 42,
    subset="training",
    validation_split=TRAIN_SPLIT,
    shuffle = True,
    image_size = (IMAGE_SIZE, IMAGE_SIZE),
    batch_size = BATCH_SIZE,
)
```

```python
val_dir = tf.keras.preprocessing.image_dataset_from_directory(
    data_dir,
    seed = 42,
    subset="validation",
    validation_split=VAL_SPLIT,
    shuffle = True,
    image_size = (IMAGE_SIZE, IMAGE_SIZE),
    batch_size = BATCH_SIZE,
)
```

```python
# Name of the classes
class_names = train_ds.class_names
print(class_names)
```

```python
# Visualize the data
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```
